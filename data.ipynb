Importing libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Exploration

```python
passPerItem_df = pd.read_csv("./passPerItemTimes.csv")
untilNoChange_df = pd.read_csv("./untilNoChangeTimes.csv")
whileNeededTimes_df = pd.read_csv("./whileNeededTimes.csv")
```

```python
passPerItem_df.head()
```

```python
passPerItem_df['Type'] = passPerItem_df['Type'].map({'best': 1, 'worst': 5, 'rand1': 2, 'rand2': 3, 'rand3': 4})
```

```python
data = pd.concat([passPerItem_df['Size'], passPerItem_df['Time'], passPerItem_df['Type']], axis=1)
fig, ax = plt.subplots()
ax.set_title("BubbleSortPassPerItem")
plt.xscale('log')
plt.ylim(0, 8e8)
ax.scatter(x='Size', y='Time', c=pd.cut(data['Type'], bins=[0, 1, 2, 3, 4, 5], labels=['#FF0000', '#FF00FF', '#800080', '#8A2BE2', '#0000FF']), data=data)
plt.show()
```

```python
summary = passPerItem_df.describe(percentiles=[.25, .5, .75])

# Extract the required values
minimum = summary.loc['min', 'Time']
first_quartile = summary.loc['25%', 'Time']
median = summary.loc['50%', 'Time']
third_quartile = summary.loc['75%', 'Time']
maximum = summary.loc['max', 'Time']

print(f"Minimum: {minimum}")
print(f"First Quartile: {first_quartile}")
print(f"Median: {median}")
print(f"Third Quartile: {third_quartile}")
print(f"Maximum: {maximum}")
```

```python
passPerItem_df['Sort'] = 1
untilNoChange_df['Sort'] = 2
whileNeededTimes_df['Sort'] = 3

df = pd.concat([passPerItem_df, untilNoChange_df, whileNeededTimes_df], ignore_index=True)
```

```python
data = pd.concat([df['Size'], df['Time'], df['Sort']], axis=1)
fig, ax = plt.subplots()
ax.set_title("BubbleSort Algorithm Comparison")
plt.xscale('log')
plt.ylim(0, 8e8)
ax.scatter(x='Size', y='Time', c=pd.cut(data['Sort'], bins=[0, 1, 2, 3], labels=['#FF0000', '#00FF00', '#0000FF']), data=data)
plt.show()
```
